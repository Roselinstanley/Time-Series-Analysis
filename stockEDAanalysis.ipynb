```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.stats import norm
from statsmodels.distributions.empirical_distribution import ECDF
import pandas as pd
```

```python
np.percentile(dataset1["Low"],89)
```

```python
def cal_pdf(dataset1):
    ax=sb.distplot(dataset1["Low"],kde=True,kde_kws={'color':'pink'},color='green')
    plt.axvline(200,color='yellow')
    plt.axvline(1000,color='yellow')
    plt.show()
    mean=dataset1["Low"].mean()
    std=dataset1["Low"].std()
    norms=norm(mean,std)
    probabilities=[norms.pdf(value) for value in range(200,1000)]
    probsum=sum(probabilities)
    probper=probsum *100
    low_values=dataset1["Low"].dropna().values.ravel()
    ecdf=ECDF(low_values)
    ecdfout=ecdf(700)
    print("The are of Low between ({},{}),({},{}),{}".format(200,1000,probsum,probper,ecdfout))
    return
```

```python
cal_pdf(dataset1)
```

```python
def std_normal(dataset1):
    mean=dataset1["Low"].mean()
    std=dataset1["Low"].std()
    values=[i for i in dataset1["Low"].dropna().values.ravel()]
    z_score=[((j-mean)/std) for j in values]
    df=pd.DataFrame({"Z_score": z_score})
    print("The Z_Score plot of plot")
    sb.distplot(z_score,kde=True)
    plt.show()
    display(df["Z_score"])
```

```python
std_normal(dataset1)
```

```python
 def frequency(dataset1):
     value_counts = dataset1["Low"].value_counts().sort_index()

     # Create the frequency table directly from value_counts
     freq = pd.DataFrame({ 
                         "Unique Values": value_counts.index,
                         "Frequency": value_counts.values
     })

     # Calculate relative frequency and CRF
     freq["Relative Frequency"] = freq["Frequency"] / freq["Frequency"].sum()
     freq["CRF"] = freq["Relative Frequency"].cumsum()

     # Display result
     display(freq)
```

```python
dataset["Low"]=dataset1["Low"].dropna().values.ravel()
frequency(dataset)
```
